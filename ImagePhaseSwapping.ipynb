```
import cv2 
import numpy as np
import matplotlib.pyplot as plt  
```

```
lena = cv2.imread('/content/lena.jpg')/255
cameraMan= cv2.imread('/content/CameraMan.png')/255 
```

```
plt.imshow(lena, cmap ='gray')
plt.title('Lena'), plt.xticks([]), plt.yticks([])
plt.show()
```

```
plt.imshow(cameraMan, cmap ='gray')
plt.title('Camera Man'), plt.xticks([]), plt.yticks([])
plt.show()
```

```
lenaFFT = np.fft.fft2(lena)
cameraManFFT = np.fft.fft2(cameraMan)

magnitudeLena = np.abs(lenaFFT)
phaseLena= np.angle(lenaFFT)

magnitudeCameraMan= np.abs(cameraManFFT)
phaseCameraMan = np.angle(cameraManFFT)

outputImag1 = magnitudeLena*np.exp(phaseCameraMan)
outputImag2 = magnitudeCameraMan * np.exp(phaseLena)

outputImag1 = np.real(np.fft.ifft2(outputImag1))
outputImag2 = np.real(np.fft.ifft2(outputImag2))
```

```
plt.imshow(outputImag1, cmap ='gray')
plt.title('Phase Swapped lena'), plt.xticks([]), plt.yticks([])
plt.show()
```

```
plt.imshow(outputImag2, cmap ='gray')
plt.title('Phase Swapped CameraMan'), plt.xticks([]), plt.yticks([])
plt.show()
```
